```python
#import necessary packages
from bs4 import BeautifulSoup
import os
import pandas as pd
```

```python
#use package os and for loop to get all the html file
for file_html in os.listdir(folder):
    #os.path.join() method in Python join one or more path components intelligently
    with open(os.path.join(folder, file_html)) as file:
#create a soup to read each file from the for loop
        soup = BeautifulSoup(file, 'lxml')
#from the soup lets extract title, user ratings and audience score
# Note: inspecting the html file is required
        title = soup.find('title')
    print(title)
    break
    
```

### ALL DONE, lets loop trough all the files and append the dictonaries to the empty list

```python
#next we find the user ratings
#lets find the audience score
#use package os and for loop to get all the html file
for file_html in os.listdir(folder):
    #os.path.join() method in Python join one or more path components intelligently
    with open(os.path.join(folder, file_html)) as file:
#create a soup to read each file from the for loop
        soup = BeautifulSoup(file, 'lxml')
#from the soup lets extract title, user ratings and audience score
# Note: inspecting the html file is required
#uses contents and slice the first element in title tags and exlude ' - Rotten Tomatoes'
        title = soup.find('title').contents[0][:-len(' - Rotten Tomatoes')]
        audience_score = soup.find('div', class_= 'critic-score meter')
        audience_score = audience_score.find_all('span')[2].contents[0]
        num_user_rating = soup.find('div', class_= 'audience-info hidden-xs superPageFontColor')
        num_user_rating = num_user_rating.find_all('div')[1].contents[2].strip().replace(',', '')
        
        df_list.append({'title': title,
                       'audience_score': int(audience_score),
                       'num_user_rating': int(num_user_rating)})
```
